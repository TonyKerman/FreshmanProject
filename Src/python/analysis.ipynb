```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
L = np.array([0.17, 0.14, 0, 0])
x = np.linspace(L[0]-L[1],L[0]+L[1],10)
theta1 = np.arccos(x/(L[0]+L[1]))/np.pi*180
theta2 = 2*theta1
theta3 =90-theta1
plt.plot(x,theta1)
plt.plot(x,theta2)
plt.plot(x,theta3)
plt.show()
```

```python
arm_length = np.array([0.17, 0.14, 0, 0],dtype=float)
servos_bis = np.array([775, 1155, 400, 700],dtype=int)
theta1 = np.arccos(arm_length[0]/(arm_length[0]+arm_length[1]))/np.pi*180
theta2 = 2*theta1
theta3 =90-theta1
target_angles=np.array([theta1,theta2,theta3])
```

```python
def angleToPos(angles):
    try:
        angles[0] = -angles[0]
        angles[1] = -angles[1]
        angles[2] = angles[2]
    except IndexError:
        pass
    if type(angles)!=np.ndarray:
        raise TypeError('type wrong')
    length = angles.shape[0]
    pos = np.zeros(length)
    if servos_bis.shape[0] >=length:
        pos=angles/0.24+servos_bis[:length]
    else:
        raise IndexError
    return  pos.astype(int)


```

```python
angleToPos(target_angles)
```
